## Estructuras de Control (Condicionales y Ciclos)
Haz un programa que ayude a automatizar procesos repetitivos y aplicar reglas de negocio (lógica condicional).

1. Crea un ciclo for que recorra tu lista de registro_ventas (puedes agregar un par de ventas manuales extra al código para hacer la prueba).
2. Dentro del ciclo, aplica un condicional if: Si el total de la venta es mayor a $100000, aplica un descuento del 10%.
3. Imprime el reporte final mostrando el producto, el total original y el total con descuento (si aplica).

```python
## Haz tu codigo aquí:
# Tupla con categorías
categorias = ("Electrónica", "Textil", "Alimentos", "Licores", "Aseo")

# Lista de ventas
registro_ventas = []

# Agregamos algunas ventas de prueba
venta = {"producto": "Computador", "categoria": "Electrónica", "precio": 2500000, "cantidad": 2, "total": 2500000 * 2}
venta1 = {"producto": "Zapatos", "categoria": "Textil", "precio": 150000, "cantidad": 3, "total": 150000 * 3}
venta2 = {"producto": "Manzana", "categoria": "Alimentos", "precio": 1200, "cantidad": 8, "total": 1200 * 8}
venta3 = {"producto": "Cerveza", "categoria": "Licores", "precio": 2500, "cantidad": 6, "total": 2500 * 6}
venta4 = {"producto": "Jabon_baño", "categoria": "Aseo", "precio": 15000, "cantidad": 4, "total": 15000 * 4}

registro_ventas.append(venta1)
registro_ventas.append(venta2)
registro_ventas.append(venta3)

# Ciclo para recorrer todas las ventas
for venta in registro_ventas:
    total_original = venta["total"]
    total_final = total_original
    descuento = 0  # porcentaje de descuento aplicado

    # Regla de negocio: 10% si supera 100000
    if total_original > 100000:
        descuento = 10
        total_final = total_original * 0.9

    # Reporte
    print(f"Producto: {venta['producto']}")
    print(f"Total original: {total_original}")
    if descuento > 0:
        print(f"Descuento aplicado: {descuento}%")
    else:
        print("Descuento aplicado: No")
    print(f"Total con descuento: {total_final}")
    print("-" * 40)
```

## Funciones y Manejo de Errores
Crea código reutilizable y a prueba de errores de usuario siguiendo estos parámetros.

1. Crea una función llamada calcular_ingreso que reciba dos parámetros: precio y cantidad.
2. Dentro de la función, utiliza un bloque try-except. Intenta multiplicar el precio por la cantidad.
3. Si ocurre un error (por ejemplo, si en lugar de un número llega un texto), el except debe atraparlo, imprimir "Error: Los valores deben ser numéricos" y retornar 0.
4. Prueba tu función enviando valores correctos y luego enviando un texto para ver cómo se maneja el error.

```python
## Haz tu codigo aquí:
def calcular_ingreso(precio, cantidad):
    try:
        ingreso = precio * cantidad
        return ingreso
    except:
        print("Error: Los valores deben ser numéricos")
        return 0

# Pedir datos al usuario
try:
    precio = float(input("Ingrese el precio del producto: "))
    cantidad = int(input("Ingrese la cantidad: "))
    
    ingreso = calcular_ingreso(precio, cantidad)
    print("Ingreso calculado:", ingreso)

except:
    print("Error: Debe ingresar valores numericos.")
```
